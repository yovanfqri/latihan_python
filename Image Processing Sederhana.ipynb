```python
import tkinter as tk
from PIL import Image
from PIL import ImageTk
from tkinter import filedialog, Menu
import cv2
import numpy as np

root = tk.Tk()
root.title('Image Processing')
root.geometry("700x600")
img = np.array([])

def select_img():
    global img, path
    path = filedialog.askopenfilename()
    if path:
        img = cv2.imread(path, cv2.COLOR_BGR2RGB)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        newImg = cv2.resize(img, (500, 500))
        
        image = Image.fromarray(newImg)
        image_tk = ImageTk.PhotoImage(image)
        
        panelSRC.configure(image=image_tk)
        panelSRC.image = image_tk
        
def blur_img():
    global img
    if img.any():
        image = cv2.GaussianBlur(img, (7,7), 0)
        img = image
        
        newImage = cv2.resize(image, (500, 500))
        
        image = Image.fromarray(newImage)
        image_tk = ImageTk.PhotoImage(image)

        panelSRC.configure(image=image_tk)
        panelSRC.image = image_tk

def edge_img():
    global img
    if img.any():
        image = cv2.Canny(img, 100,200)

        newImage = cv2.resize(image, (500, 500))
        
        image = Image.fromarray(newImage)
        image_tk = ImageTk.PhotoImage(image)

        panelSRC.configure(image=image_tk)
        panelSRC.image = image_tk
        
def gray_img():
    global img
    if img.any():
        image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        newImage = cv2.resize(image, (500, 500))
        
        image = Image.fromarray(newImage)
        image_tk = ImageTk.PhotoImage(image)

        panelSRC.configure(image=image_tk)
        panelSRC.image = image_tk

def contrast_img():
    global img
    if img.any():
        image = cv2.addWeighted(img, 2.5, np.zeros(img.shape, img.dtype), 0, 0)
        img = image

        newImage = cv2.resize(image, (500, 500))
        
        image = Image.fromarray(newImage)
        image_tk = ImageTk.PhotoImage(image)

        panelSRC.configure(image=image_tk)
        panelSRC.image = image_tk

def reset_img():
    global img
    if img.any():
        img = cv2.imread(path, cv2.COLOR_BGR2RGB)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        newImg = cv2.resize(img, (500, 500))
        
        image = Image.fromarray(newImg)
        image_tk = ImageTk.PhotoImage(image)

        panelSRC.configure(image=image_tk)
        panelSRC.image = image_tk

w = tk.Label(root, text="Image Processing Dengan Python", font="Arial 18 bold", fg="red", justify="center")
w.pack()

#Menu Bar
menubar = Menu(root)
root.config(menu=menubar) 

file_menu = Menu(menubar, tearoff=False)
menubar.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="Choose Image", command=select_img)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.destroy)

edit_menu = Menu(menubar, tearoff=False)
menubar.add_cascade(label="Edit", menu=edit_menu)
edit_menu.add_command(label="Reset", command=reset_img)

#Buttons
edgebtn = tk.Button(root, text="Edge Detection", command=edge_img, width=12)
edgebtn.place(x=120, y=50, anchor="nw")

blurbtn = tk.Button(root, text='Blur Image', command=blur_img, width=12)
blurbtn.place(x=250, y=50, anchor="nw")

graybtn = tk.Button(root, text='Grayscale Image', command=gray_img, width=12)
graybtn.place(x=370, y=50, anchor="nw")

contimg = tk.Button(root, text="Contrast Image", command=contrast_img, width=12)
contimg.place(x=500, y=50, anchor="nw")

panelSRC = tk.Label(root)
panelSRC.pack(side="bottom", padx=10, pady=10)

root.mainloop()
```
